```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading dataframe 1
data1 = pd.read_csv('/content/drive/MyDrive/Robotic Lab/data1.csv', sep = ';')
data1
```

```python
#Loading dataframe 2
data2 = pd.read_csv('/content/drive/MyDrive/Robotic Lab/data2.csv', sep = ';')
data2
```

```python
#Loading dataframe 3
data3 = pd.read_csv('/content/drive/MyDrive/Robotic Lab/data3.csv',sep = ';')
data3
```

```python
#Merging the three dataframes
#Merging the first two datas
data12 = pd.merge(data1, data2, how = 'right')
data12
```

```python
#Merging with the last data frame
data = pd.concat([data12, data3])
data
```

```python
data.describe()
```

```python
data.dtypes
```

```python
#Cleaning the data
data.isnull() #checking it has a value or not
```

```python
data.drop_duplicates(inplace = True) #dropping duplicates data
data
```

```python
data.dropna(inplace = True) #dropping out null values
data
```

```python
#Removing Impossible Values
finaldata = data.loc[(data['Age at enrollment'] >= 17) & (data['Age at enrollment'] <= 70)]
finaldata
```

```python
finaldata.Target.astype('string')
```

```python
final_data = finaldata.apply(lambda x: x.str.strip() if x.dtype == "string" else x) #removing extra white space
final_data
```

```python
#Remove white spaces 
# check for leading/trailing white spaces
print(final_data['Target'].str.startswith(' ').sum())
print(final_data['Target'].str.endswith(' ').sum())

# remove leading/trailing white spaces
final_data['Target'] = final_data['Target'].str.strip()
```

```python
#Check for misspelled words
!pip install pyspellchecker
from spellchecker import SpellChecker

# Create a spell checker object
spell = SpellChecker()

# Load the target column as a pandas Series
target_col = final_data['Target']

# Loop through each word in the Series and check if it is misspelled
misspelled = spell.unknown(target_col)

# Print the misspelled words
print(misspelled)
```

```python
misspelled_words = {'drpout', 'dropot', 'enroled', 'grduate'}

# replace misspelled words in target column
final_data['Target'] = final_data['Target'].apply(lambda x: 'Dropout' if x.lower() == 'drpout' else x)
final_data['Target'] = final_data['Target'].apply(lambda x: 'Graduate' if x.lower() == 'grduate' else x)
final_data['Target'] = final_data['Target'].apply(lambda x: 'Enrolled' if x.lower() in ['enroled', 'enrolled'] else x)
final_data['Target'] = final_data['Target'].apply(lambda x: 'Dropout' if x.lower() == 'dropot' else x)

# check unique values in target column
print(final_data['Target'].unique())
```

```python
#Choosing two columns of categorical values
print('Categorical Values')
print('Target:')
print(final_data.Target)
print('Nationality:')
print(final_data.Nationality)
```

```python
#Choosing two columns of numerical values
print('Numerical Values')
print('Admission Grade:')
print(final_data['Admission grade'])
print('GDP:')
print(final_data.GDP)
```

```python
#Choosing two columns of categorical value and numerical value
plt.boxplot([final_data[final_data['Target'] == 'Dropout']['Admission grade'],
             final_data[final_data['Target'] == 'Enrolled']['Admission grade'],
             final_data[final_data['Target'] == 'Graduate']['Admission grade']])
plt.xticks([1, 2, 3], ['Dropout', 'Enrolled', 'Graduate'])
plt.ylabel('Admission grade')
plt.title('Distribution of Admission grade for each Target category')
plt.show()
```

```python
#Previous Qualification Grade and Application Order
x =final_data['Nationality']
y=final_data['GDP']
plt.scatter(x,y, c = 'r', cmap = 'viridis') 
plt.title('GDP vs Nationality')
plt.show()
```

```python
#Histogram
age_enrollment = final_data[['Age at enrollment', 'Target']]
# Filter out taget columns which is not = to graduate, we only want to know data of the ones that graduated
age_enrollment = age_enrollment[age_enrollment['Target'] == 'Graduate']
plt.hist(age_enrollment['Age at enrollment'], bins=20, range=(15,50))
plt.xlabel('Age at enrollment')
plt.ylabel('Number of students')
plt.title('Distribution of age at enrollment for graduates')
plt.show()
```

```python
#Choosing three pairs of columns and exploring the relationship
#Age at Enrollment and Admission Grade
final_data.plot(x = final_data.columns[20], y = [final_data.columns[13]] ,kind = 'scatter')
plt.title('The scatter plot illustrates the relationship between enrollment age and admission grade')
plt.xlabel('Enrollment Age')
plt.ylabel('Admission Grade')
plt.show()
```

```python
#Adimssion Grade and Previous Qualification (Grade)
x1=final_data['Admission grade']
y1=final_data['Previous qualification (grade)']
plt.scatter(x1,y1, c = 'r', cmap = 'viridis') 
plt.title('Previous vs Admission Grades')
plt.show()
```

```python
#Gender and Target 
gender_counts = final_data['Gender'].value_counts()
target_counts = final_data['Target'].value_counts()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
ax1.set_title('Gender Distribution')
ax2.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%')
ax2.set_title('Target Distribution')
plt.show()
```
